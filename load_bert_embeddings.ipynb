## Example code to load BERT word embedding files 

```python
import pickle
import pandas as pd
chunk_len = 512
fname = 'data/distilBert_embeddings_olap_' + str(chunk_len) + '.pkl'

with open(fname, 'rb') as fp:
    df_olap = pickle.load(fp)
```
